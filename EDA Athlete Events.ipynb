# Identifying the variables

```python
numerical_vars = ['Age','Height', 'Weight','Year']
categorical_vars = ['Sex', 'Team', 'NOC', 'Games', 'Season', 'Sport', 'Event','Medal',]
```

```python
df = df[numerical_vars+categorical_vars]
```

```python
df[numerical_vars].describe()
```

```python
df[numerical_vars].hist(edgecolor='black', bins=15, figsize=(14, 5), layout = (2,4));
```

```python
df['Sex'].value_counts().plot(kind='bar', title='Sex');
```

```python
df['Medal'].value_counts().plot(kind='bar', title='Medal');
```

```python
df['Season'].value_counts().plot(kind='bar', title='Season');
```

From the above analysis we can make out that the Age group of Athlete was more between 20-40 yrs.
There are more Male than Female participants.
More games were held for Summer than Winter.
